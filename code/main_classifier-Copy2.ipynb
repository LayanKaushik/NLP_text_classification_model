#### CNN using keras

```python
def cnn_model(input_length, num_filters, kernel_size, embedding_dim):
    model = Sequential()
    model.add(Embedding(input_dim=input_length, output_dim=embedding_dim, input_length=input_length))
    model.add(Conv1D(filters=num_filters, kernel_size=kernel_size, activation='relu'))
    model.add(GlobalMaxPooling1D())
    model.add(Dense(10, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(1, activation='sigmoid'))  # Binary classification
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Define model parameters
num_filters = 64
kernel_size = 3
embedding_dim = 100

# Using Tf-Idf
print_separator("CNN with Tf-Idf")
input_length = X_train_tfidf.shape[1]
model_tfidf = cnn_model(input_length, num_filters, kernel_size, embedding_dim)
model_tfidf.fit(X_train_tfidf, y_train, epochs=10, batch_size=32, validation_split=0.1)
loss_tfidf, accuracy_tfidf = model_tfidf.evaluate(X_test_tfidf, y_test)
print(f'Tf-Idf Test Loss: {loss_tfidf:.4f}, Test Accuracy: {accuracy_tfidf:.4f}')

# Using Word2Vec
print_separator("CNN with Word2Vec")
model_word2vec = cnn_model(X_train_word2vec.shape[1], num_filters, kernel_size, embedding_dim)
model_word2vec.fit(X_train_word2vec, y_train, epochs=10, batch_size=32, validation_split=0.1)
loss_word2vec, accuracy_word2vec = model_word2vec.evaluate(X_test_word2vec, y_test)
print(f'Word2Vec Test Loss: {loss_word2vec:.4f}, Test Accuracy: {accuracy_word2vec:.4f}')

# Using GloVe
print_separator("CNN with GloVe")
model_glove = cnn_model(X_train_glove.shape[1], num_filters, kernel_size, embedding_dim)
model_glove.fit(X_train_glove, y_train, epochs=10, batch_size=32, validation_split=0.1)
loss_glove, accuracy_glove = model_glove.evaluate(X_test_glove, y_test)
print(f'GloVe Test Loss: {loss_glove:.4f}, Test Accuracy: {accuracy_glove:.4f}')

# # Using BERT
# print_separator("CNN with BERT")
# model_bert = cnn_model(X_train_bert.shape[1], num_filters, kernel_size, embedding_dim)
# model_bert.fit(X_train_bert, y_train, epochs=10, batch_size=32, validation_split=0.1)
# loss_bert, accuracy_bert = model_bert.evaluate(X_test_bert, y_test)
# print(f'BERT Test Loss: {loss_bert:.4f}, Test Accuracy: {accuracy_bert:.4f}')
```

#### CNN Trial

```python
def cnn_model2(input_length, num_filters, kernel_sizes, embedding_dim, dropout_rate=0.5):
    model = Sequential()
    model.add(Embedding(input_dim=input_length, output_dim=embedding_dim, input_length=input_length))

    for kernel_size in kernel_sizes:
        model.add(Conv1D(filters=num_filters, kernel_size=kernel_size, activation='relu'))
        model.add(GlobalMaxPooling1D())

    model.add(Dense(64, activation='relu'))
    model.add(Dropout(dropout_rate))
    model.add(Dense(32, activation='relu'))
    model.add(Dropout(dropout_rate))
    model.add(Dense(1, activation='sigmoid'))  # Binary classification

    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Model parameters
num_filters = 64
kernel_sizes = [3, 4, 5]
embedding_dim = 100
dropout_rate = 0.5

# Using Tf-Idf
print_separator("CNN with Multiple Kernels - Tf-Idf")
input_length_tfidf = X_train_tfidf.shape[1]
model_tfidf = cnn_model2(input_length_tfidf, num_filters, kernel_sizes, embedding_dim, dropout_rate)
model_tfidf.fit(X_train_tfidf, y_train, epochs=10, batch_size=32, validation_split=0.1)
loss_tfidf, accuracy_tfidf = model_tfidf.evaluate(X_test_tfidf, y_test)
print(f'Tf-Idf Test Loss: {loss_tfidf:.4f}, Test Accuracy: {accuracy_tfidf:.4f}')

# Using Word2Vec
print_separator("CNN with Multiple Kernels - Word2Vec")
input_length_word2vec = X_train_word2vec.shape[1]
model_word2vec = cnn_model2(input_length_word2vec, num_filters, kernel_sizes, embedding_dim, dropout_rate)
model_word2vec.fit(X_train_word2vec, y_train, epochs=10, batch_size=32, validation_split=0.1)
loss_word2vec, accuracy_word2vec = model_word2vec.evaluate(X_test_word2vec, y_test)
print(f'Word2Vec Test Loss: {loss_word2vec:.4f}, Test Accuracy: {accuracy_word2vec:.4f}')

# Using GloVe
print_separator("CNN with Multiple Kernels - GloVe")
input_length_glove = X_train_glove.shape[1]
model_glove = cnn_model2(input_length_glove, num_filters, kernel_sizes, embedding_dim, dropout_rate)
model_glove.fit(X_train_glove, y_train, epochs=10, batch_size=32, validation_split=0.1)
loss_glove, accuracy_glove = model_glove.evaluate(X_test_glove, y_test)
print(f'GloVe Test Loss: {loss_glove:.4f}, Test Accuracy: {accuracy_glove:.4f}')

# # Using BERT
# print_separator("CNN with Multiple Kernels - BERT")
# input_length_bert = X_train_bert.shape[1]
# model_bert = cnn_model2(input_length_bert, num_filters, kernel_sizes, embedding_dim, dropout_rate)
# model_bert.fit(X_train_bert, y_train, epochs=10, batch_size=32, validation_split=0.1)
# loss_bert, accuracy_bert = model_bert.evaluate(X_test_bert, y_test)
# print(f'BERT Test Loss: {loss_bert:.4f}, Test Accuracy: {accuracy_bert:.4f}')
```
